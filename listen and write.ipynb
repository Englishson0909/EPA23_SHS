```python
#@markdown Define ktts(), etts()
def etts(txt):
  tts = gTTS(txt, lang='en')
  tts.save("E-audio.mp3")
  return Audio("/content/E-audio.mp3")

def ktts(txt):
  tts = gTTS(txt, lang='ko')
  tts.save("K-audio.mp3")
  return Audio("/content/K-audio.mp3")
```
